```python
# Univariate Linear Regression Without Libraries
X = [1,2,3,4]
Y = [2,4,6,8]
a1 = 0
a0 = 0
```

```python
learning_rate = 0.01
epochs =1000
n = len(X)
a0_value=[]
a1_value=[]
```

```python
for epoch in range(epochs):
    Y_pred = [(a1 * x + a0) for x in X]
    da1 = (-2/n) * sum((Y[i] - Y_pred[i]) * X[i] for i in range(n))
    da0 = (-2/n) * sum((Y[i] - Y_pred[i]) for i in range(n))
    a1 -= learning_rate * da1
    a0 -= learning_rate * da0
```

```python
print(f"Final Model a1:{a1:.2f}")
print(f"Final Model a0:{a0:.2f}")
print(f"Final Model: Y = {a1:.2f}X + {a0:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(a0_value, label='a0')
plt.plot(a1_value, label='a1')
plt.xlabel("Epochs")
plt.ylabel("Value")
plt.title("Convergence of a0 and a1")
plt.legend()
plt.show()
```
